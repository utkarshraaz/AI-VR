```python
#import useful libraries
import pandas as pd
import re
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Initialize the sentence Transformer model for sentence embedding.
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
#Read the CSV File containing Questions and answer related to Pine Tree.
Tree_data = pd.read_csv("pine_data.csv")
Tree_data.head()
```

```python
# Prepare data for vectorization
known_questions = list(Tree_data['Question'])
known_answers = list(Tree_data['Answer'])

# Create embeddings for known questions
known_question_embeddings = model.encode(known_questions)
```

```python
# Function to fetch the most similar answer to the asked questions.
def find_best_answer(user_query):
    # Create embedding for user query
    query_embedding = model.encode([user_query])

    # Calculate cosine similarities between user query and known questions
    similarities = cosine_similarity(query_embedding, known_question_embeddings)

    # Find the best matching known question
    best_match_idx = np.argmax(similarities)

    #Checking if the asked question is atleast 70 % related to best answer.
    if similarities[0][best_match_idx]>=0.7:
      return known_answers[best_match_idx]
    else:
      return "Sorry, I'm still learning and I don't have an answer to that question! Could you please rephrase your question?"
```

```python
# function to start chat.
def chat():
    print("Hi, ask me anything about Pine trees!")
    while True:
        user_input = input()
        if user_input.lower() == 'quit':
            print("Thank you!")
            break
        print(find_best_answer(user_input))

# Start the chat
chat()
```
